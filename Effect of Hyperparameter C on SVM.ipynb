# Generating Linearly Separable Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
x,y = make_blobs(centers=2,cluster_std=2.5,random_state=1)
sns.scatterplot(x[:,0],x[:,1],hue=y);
```

# Using scikit-learn SVC to classify the dataset

```python
from sklearn.svm import SVC
clf = SVC(C=1.0,kernel='linear')
clf.fit(x,y)
```

# Calculating coefficient and intercept of the hyperplane and marginal hyperplane.

```python
# m is the slope of the hyperplane and b is the intercept of hyperplane
# where b1 and b2 is the intercept of marginal hyperplane.
m = -(clf.coef_[0][0]/clf.coef_[0][1])
b = -(clf.intercept_[0]/clf.coef_[0][1])
b1 = -((clf.intercept_[0]-1)/clf.coef_[0][1])
b2 = -((clf.intercept_[0]+1)/clf.coef_[0][1])
```

# Plotting the hyperplane and marginal hyperplane for the given dataset.

```python
x1 = np.linspace(-15,5,20)
f1 = m*x1 + b
f2 = m*x1 + b1
f3 = m*x1 + b2

sns.lineplot(x1,f1)
sns.lineplot(x1,f2)
sns.lineplot(x1,f3)
sns.scatterplot(x[:,0],x[:,1],hue=y);
```

# Generating dataset which are not completely linearly separable

```python
x,y = make_blobs(centers=2,cluster_std=3,random_state=1)
sns.scatterplot(x[:,0],x[:,1],hue=y);
```

# Classify the dataset using SVM with lower value of hyperparameter C=0.01

```python
clf1 = SVC(C=10,kernel='linear')
clf1.fit(x,y)

m = -(clf1.coef_[0][0]/clf1.coef_[0][1])
b = -(clf1.intercept_[0]/clf1.coef_[0][1])
b1 = -((clf1.intercept_[0]-1)/clf1.coef_[0][1])
b2 = -((clf1.intercept_[0]+1)/clf1.coef_[0][1])

x1 = np.linspace(-15,5,20)
f1 = m*x1 + b
f2 = m*x1 + b1
f3 = m*x1 + b2

sns.lineplot(x1,f1)
sns.lineplot(x1,f2)
sns.lineplot(x1,f3)
sns.scatterplot(x[:,0],x[:,1],hue=y);
```

# Classify the dataset using SVM with higher value of hyperparameter C=100

```python
clf2 = SVC(C=0.01,kernel='linear')
clf2.fit(x,y)

m = -(clf2.coef_[0][0]/clf2.coef_[0][1])
b = -(clf2.intercept_[0]/clf2.coef_[0][1])
b1 = -((clf2.intercept_[0]-1)/clf2.coef_[0][1])
b2 = -((clf2.intercept_[0]+1)/clf2.coef_[0][1])

x1 = np.linspace(-15,5,20)
f1 = m*x1 + b
f2 = m*x1 + b1
f3 = m*x1 + b2

sns.lineplot(x1,f1)
sns.lineplot(x1,f2)
sns.lineplot(x1,f3)
sns.scatterplot(x[:,0],x[:,1],hue=y);
```

From above Analysis i conclude that The hyperparameter C is important to decide the mis-clssification error.                                
If value of C is small then SVM give more importance to mis-classification error and try very hard to correctly classify each data point.                                                                 
But, If C value is large the SVM gives more importance to margin and increase the margin on cost of some missclssification.
